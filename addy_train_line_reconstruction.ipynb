# Major discoveries
ID_part2 is not actually the time of snapshot, but the time at which the train departed from the first station.

The data was pulled every hour, 2 minutes after the full hour.

I believe that the pulling software somehow did not account for many stations. From the examples above, we can see that there are NO data points where the station is Zöberitz or Niemberg. However, Schönebeck (Elbe) is represented both in the stations and in the path as the last element.

This indicates either some mistakes in the pulling or the pulling code not accounting for certain times or stations. Stations might be based on size.

```python
data.category.unique()
```

This seems to confirm my suspicions -- stations with category 6 or 7 are not included in the data. So if the line is rural (goes through small / very small stations), we don't have data points for those.
