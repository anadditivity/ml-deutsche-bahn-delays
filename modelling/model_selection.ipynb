One thing we should keep in mind is that our dataset is heavily imbalanced with only ~5% of our data set being delayed. We therefore should use something like the f-score or just the recall as our measurement for how well a given model has performed. 

Due to the amount of data it is also hard for us too say upfront which parameters are going to perform well and which won't. Since we have a lot of data points available we will opt to use k-fold cross validation to find the best parameter combination for the model we choose.

In order to deal with the imbalance in our dataset we will just sample the same amount of trains that were on time, as we have trains that were delayed and directly put the other trains that were on time into the test set.

## Decision Tree Classifier

Decision trees have the advantage that they are easy to understand and can be interpreted easily as well. However they are often a bit simple and suffer more easily from overfitting than other models.

## Random Forest Classifier

Random Forest Classifiers are probably a better choice than one single decision tree, because they are less prone too overfitting if enough data points are available, which is given in our case with 2061357 datapoints.

## Gradient Boosting

Gradient Boosting is most likely a better fit than a random forest classifier. The advantages of this model are that it will probably be able to achieve a high accuracy. The problem is that we will most likely need to do a lot of cross validation for the hyperparameters and this will take a lot of time due to the nature of the model.

## K-Nearest-Neighbor 

KNN is also a possible choice. Since we have significantly more data points than features (~2*10⁷ >> 20), it is unlikely that the course of dimensionality is going to be a big problem in our case. We will most likely use only ~10% of the datapoints available for training but these are still way more datapoints. One problem we might encounter is that we should probably switch to one-hot-encoding for different nominal values too be able too use them. This will blow up the amount of features that we have but it will most likely still be less than the amount of data points we have. But this is definetly something to keep in mind. 

If we see that we don't have enough datapoints we can still use PCA to reduce dimensionality again.

## Neural Network

A neural network is possibly a very accurate solution as we have a lot of datapoints and can therefore train the model very long. However we do think it is not necessary tooo train a neural network as other more efficient options are also available that will probably reach almost the same accuracy as a neural network.

## Support Vector Machine

We don't think that SVMs are a suitable approach to our problem because if we use one hot encoding we have a lot of dimensions and a lot of data points which will make the use of a SVM very complex.

## Conclusion

We will run some smaller tests on the efficiency and accuracy/f-score of a gradient boosting model. In case of promising results we will continue too use them. Otherwise we will choose to go for a KNN model since we deem it to be the next best fit because we have a lot of data available and it is way less time consuming allowing us to try out more hyperparameters. In case this model does not provide a satisfying f-score either, we will try to apply PCA or go for a neural network instead.
